    + **Indicator columns(with value True/False) about keywords Python, SQL, AWS, RESTFUL, Machine learning, Deep Learning, Text Mining, NLP, SAS, Tableau, Sagemaker, TensorFlow, Spark**

```python
import re
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

```python
jobs_df = pd.DataFrame(columns = ['title' , 'company_name' , 'location' , 'summary' , 'python' , 'sql' , 'aws' , 'restful' , 'machine_learning' , 'deep_learning' , 'text_mining' , 'nlp' , 'sas' , 'tableau' , 'sagemaker' , 'tensorflow' , 'spark']) 

#jobs_df
```

```python
base = 'Can no longer go straight to indeed'
page_num = 10
#newUrl = base + 'start=' + str(page_num)
req = requests.get(base)
#req = Request(base , headers={'User-Agent': 'Mozilla/5.0'})
#page = urlopen(req).read()
indeed = 'https://www.indeed.com'

while page_num < 100:
    
    #print('starting loop')
    soup = BeautifulSoup(req.text, "html.parser")
    print('--------------------------------------------------------------------------------------')
    #print(req.text)
    #soup = BeautifulSoup(page , "html.parser")
    jobs = soup.find_all('a' , attrs={'class': 'tapItem'})
    #print(jobs)
    print(len(jobs))
    for job in jobs:
        #print('made it')
        tagSource = job.get("href")
        title = job.find('span' , class_=lambda x: x != 'label')
        compName = job.find('span' , attrs={'class': 'companyName'})
        location = job.find('div' , attrs={'class': 'companyLocation'})
        summary = job.find('div' , attrs={'class': 'job-snippet'})
        
        
        listingUrl = indeed + tagSource
        listing = requests.get(listingUrl)
        #print(listingUrl)
        listingSoup = BeautifulSoup(listing.text , "html.parser")
        #print(listingSoup.text)
        description = listingSoup.find('div' , attrs={'id': 'jobDescriptionText'})
        python, sql, aws, restful, ml, dl, textMining , nlp , sas , tableau , sagemaker , tensorflow, spark = False , False , False , False , False , False , False , False , False , False , False , False , False
        
        if re.search( 'python', listing.text , re.IGNORECASE):
            python = True
        #print(listing.text)
        #input()
        if re.search( 'sql', listing.text , re.IGNORECASE):
            sql = True
        if re.search( 'aws', listing.text , re.IGNORECASE):
            aws = True
        if re.search( 'restful', listing.text , re.IGNORECASE):
            restful = True
        if re.search( 'Machine learning', listing.text , re.IGNORECASE):
            ml = True
        if re.search( 'Deep learning', listing.text , re.IGNORECASE):
            dl = True
        if re.search( 'Text Mining', listing.text , re.IGNORECASE):
            textMining = True
        if re.search( 'NLP', listing.text , re.IGNORECASE):
            nlp = True
        if re.search( 'SAS', listing.text , re.IGNORECASE):
            sas = True
        if re.search( 'Tableau', listing.text , re.IGNORECASE):
            tableau = True
        if re.search( 'Sagemaker', listing.text , re.IGNORECASE):
            sagemaker = True
        if re.search( 'TensorFlow', listing.text , re.IGNORECASE):
            tensorflow = True
        if re.search( 'Spark', listing.text , re.IGNORECASE):
            spark = True
        
        jobs_df = jobs_df.append({'title': title.text , 'company_name': compName.text , 'location': location.text , 'summary': summary.text , 'python': python,
                                 'sql': sql , 'aws': aws , 'restful': restful , 'machine_learning': ml , 'deep_learning': dl , 'text_mining': textMining , 'nlp': nlp,
                                 'sas': sas , 'tableau': tableau , 'sagemaker': sagemaker , 'tensorflow': tensorflow , 'spark': spark}, ignore_index=True)
        
        #adding wait because I keep getting locked out
        time.sleep(10)
#     del soup
#     del jobs
#     del req

    newUrl = base + '&start=' + str(page_num)

    req = requests.get(newUrl)
    #req = Request(newUrl , headers={'User-Agent': 'Mozilla/5.0'})
    #page = urlopen(req).read()
    
    page_num += 10
    #print('temp end')
    #adding wait because I keep getting locked out
    time.sleep(30)
    
    
jobs_df
        
    
```
